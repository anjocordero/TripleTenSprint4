# Introduction

In this notebook, we'll be exploring the provided car advertisement dataset to flex our data analytics muscles and apply them to a web application. We'll start by importing `pandas` and `plotly_express` to help work with our data.

```python
import pandas as pd
import plotly_express as px
```

Let's read our csv file into a DataFrame and take a high-level look at it.

```python
df = pd.read_csv("../vehicles_us.csv")
df.info()
df.describe(include='all')
```

We can immediately see that some of the columns have missing values, with fewer than the maximum of 51525 values. Depending on what we want to analyze, we'll need to decide how we want to treat these missing values later on. We can also take a look at a sample of the data:

```python
df.head(10)
```

The column `is_4wd` seems to use the number 1.0 in place of a Boolean, and the `model` column seems to combine the manufacturer and make into one column. We'll grab the first word to isolate the manufacturer.

## Fixing Data

Before we make the change, let's make sure that the column behaves how we think by checking the unique values in the column.

```python
df['is_4wd'].unique()
```

Assuming that 1.0 means `True` and nan is `False`, we can then replace the values.

```python
df['is_4wd'] = df['is_4wd'] == 1.0
df.head(10)
```

Now we can take the first word of the `model` column and enrich the data with a new `manufacturer` column.

```python
df['manufacturer'] = df['model'].apply(lambda x: x.split()[0])
manufacturers = sorted(df['manufacturer'].unique())
df.head()
```

# Exploratory Data Analysis

We can now start examining the dataset to try to find any interesting associations. A couple ideas that come to mind are the relationship between `model_year` and `price`, `condition` vs. `days_listed`, or histograms showing the distribution of `odometer` readings, `model_year`, or `days_listed`.

```python
px.histogram(df['days_listed'])
```

```python
px.histogram(
    df['price'],
    range_x=[0, 75000],
    nbins=700
    )
```

Let's also create a scatter plot showing price per manufacturer, so that we can see any major differences between them.

```python
px.scatter(
    df,
    x = 'price',
    y = 'model_year',
    labels = {
        'price':'Price',
        'model_year':'Model Year',
        'manufacturer':'Manufacturer'
    },
    color = 'manufacturer',
    title = 'Vehicle Price vs. Model Year by Manufacturer',
    opacity = 0.66
)
```
